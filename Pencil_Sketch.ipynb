## importing the library

```python
import cv2
import matplotlib.pyplot as plt 
```

## read the image

```python
read_img=cv2.imread('F:\lion-babys.jpg')
img = cv2.cvtColor(read_img,cv2.COLOR_BGR2RGB)
plt.figure(figsize=(8,5))
plt.imshow(img)
plt.axis("off")
plt.title("This is Original Image")
plt.show()
```

## converting the image into grayscale

```python
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.figure(figsize=(8,5))
plt.imshow(img_gray,cmap="gray")
plt.axis("off")
plt.title("GrayScale Image")
plt.show()
```

## inverting the image

```python
img_invert = cv2.bitwise_not(img_gray)
plt.figure(figsize=(8,5))
plt.imshow(img_invert,cmap="gray")
plt.axis("off")
plt.title("Inverted Image")
plt.show()
```

## blurring the image

```python
img_smoothing = cv2.GaussianBlur(img_invert, (5,5),sigmaX=0,sigmaY=0)
plt.figure(figsize=(8,5))
plt.imshow(img_smoothing,cmap="gray")
plt.axis("off")
plt.title("Smoothen Image")
plt.show()
```

## converting image to pencil sketche

```python
final = cv2.divide(img_gray, 255 - img_smoothing,scale=255)
plt.figure(figsize=(8,5))
plt.imshow(final,cmap="gray")
plt.axis("off")
plt.title("Final Sketch Image")
plt.show()
```

## display all image

```python
plt.figure(figsize=(20,20))
plt.subplot(1,5,1)
plt.imshow(img)
plt.axis("off")
plt.title("Original Image")
plt.subplot(1,5,2)
plt.imshow(img_gray,cmap="gray")
plt.axis("off")
plt.title("GrayScale Image")
plt.subplot(1,5,3)
plt.imshow(img_invert,cmap="gray")
plt.axis("off")
plt.title("Inverted Image")
plt.subplot(1,5,4)
plt.imshow(img_smoothing,cmap="gray")
plt.axis("off")
plt.title("Smoothen Image")
plt.subplot(1,5,5)
plt.imshow(final,cmap="gray")
plt.axis("off")
plt.title("Final Sketch Image")
plt.show()
```
